```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

def scaler(df):
    max_min_scaler = lambda x : (x-np.min(x))/(np.max(x)-np.min(x))
    dk=df.apply(max_min_scaler)
    return dk

# 
def ShowGRAHeatMap(df):
    colormap = plt.cm.hsv
    ylabels = df.columns.values.tolist()
    f, ax = plt.subplots(figsize=(15, 15))
    ax.set_title('Heat map')
    # mask
    mask = np.zeros_like(df)
    # mask[np.triu_indices_from(mask)] = True  # np.triu_indices 
    # mask[df<0.0] = True
 
    with sns.axes_style("white"):
        sns.heatmap(df,
                    cmap="hot_r",
                    annot=False,
                    mask=mask,
                    )
    plt.show()

# path
np_raw = np.load('corr_graph.npy')
df = pd.DataFrame(np_raw[0,0]).abs()
# max min scaler
# df = scaler(df)
# show
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    ShowGRAHeatMap(df)
```
